```python
# example-1a: conversion in one compartment
import tellurium as te
te.setDefaultPlottingEngine('matplotlib')

model = """
model test
    compartment C1;
    C1 = 1.0;
    species S1, S2;

    S1 = 10.0;
    S2 = 0.0;
    S1 in C1; S2 in C1;
    J1: S1 -> S2; k1*S1;

    k1 = 1.0;
end
"""
# load models
r = te.loada(model)

# simulate from 0 to 50 with 100 steps
r.simulate(0, 20, 100)
# plot the simulation
r.plot()
```

```python
# example-1b: conversion in two compartment
import tellurium as te
te.setDefaultPlottingEngine('matplotlib')

model = """
model test
    compartment C1, C2;
    C1 = 1.0;
    C2 = 5.0; # try changing C1, C2 size
    species S1, S2;

    S1 = 10.0;
    S2 = 0.0;
    S1 in C1; S2 in C2;
    J1: S1 -> S2; k1*S1;

    k1 = 1.0;
end
"""
# load models
r = te.loada(model)

# simulate from 0 to 50 with 100 steps
r.simulate(0, 50, 10)
# plot the simulation
r.plot()
```

```python
# example-1: conversion in one compartment
import tellurium as te
te.setDefaultPlottingEngine('matplotlib')

model = """
model test
    compartment C1;
    C1 = 1.0;
    species S1, S2, S3;

    S1 = 10.0;
    S2 = 0.0;
    S1 in C1; S2 in C1;
    J1: S1 -> S2; k1*S1;
    J2: S2 -> S3; k2*S2;

    k1 = 1.0;
    k2 = 1.0;
end
"""
# load models
r = te.loada(model)

# simulate from 0 to 50 with 100 steps
r.simulate(0, 20, 100)
# plot the simulation
r.plot()
```
