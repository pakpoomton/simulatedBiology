```python
import tellurium as te
import numpy as np
import matplotlib.pyplot as plt

# Load model from SBML
r = te.loadSBMLModel('BIOMD0000000010.xml')
result = r.simulate(0, 3000, 5000)

#get column name
columnName = result.colnames
print("column name: " + str(columnName))
#see output dimension
print("output dimension: " + str(result.shape))
print("column 1: " + str(result[:,1]))
print("row 1: " + str(result[:,1]))



# extract timecourse value of each chemical species
time = result['time'] #time points in simulation
MKKK = result['[MKKK]']
MKKK_P = result['[MKKK_P]']
MKK = result['[MKK]']
MKK_P = result['[MKK_P]']
MAPK = result['[MAPK]']
MAPK_P = result['[MAPK_P]']
MAPK_PP = result['[MAPK_PP]']

plt.subplots(figsize=(12,6 )) # set figure size

p1 = plt.plot(time, MKKK, color = 'k', label = "MKKK")
p2 = plt.plot(time, MKKK_P, linestyle = 'dotted', color = 'k', label = "MKKK_P")
p3 = plt.plot(time, MKK, color = 'b', label = "MKK")
p4 = plt.plot(time, MKK_P, linestyle = 'dotted', color = 'b', label = "MKK_P")
p5 = plt.plot(time, MAPK, color = 'r', label = "MAPK")
p6 = plt.plot(time, MAPK_P, linestyle = 'dotted', color = 'r', label = "MAPK_P")
p7 = plt.plot(time, MAPK_PP, linestyle = 'dashed', color = 'r', label = "MAPK_PP")

plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1))

plt.xlabel('time', fontsize = 16)
plt.ylabel('conc.', fontsize = 16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```
